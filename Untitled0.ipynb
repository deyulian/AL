```python
import numpy as np
import math
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score, mean_squared_error
from xgboost import XGBRegressor
from sklearn.ensemble import RandomForestRegressor

# === Test functions ===
def forrester(x):
    x = np.array(x)
    return (6*x - 2)**2 * np.sin(12*x - 4)

def rosenbrock5(x):
    x = np.array(x)
    return np.sum(100*(x[1:] - x[:-1]**2)**2 + (x[:-1]-1)**2)

def sobol_g(x, a=None):
    x = np.array(x)
    if a is None:
        a = np.ones(x.size)
    return np.prod((np.abs(4*x-2)+a)/(1+a))

# === Utilities ===
def featurize(X):
    return np.hstack([X, np.sin(2*np.pi*X), np.cos(2*np.pi*X)])

def simulate_samples(samples, func):
    X = np.array(samples)
    y = np.array([func(x) for x in X])
    return X, y

def train_model(X, y, params):
    Xf = featurize(X)
    scaler = StandardScaler()
    Xs = scaler.fit_transform(Xf)
    #model = XGBRegressor(**params)

    #RFR
    y = y.ravel()
    model = RandomForestRegressor(
      n_estimators=100,
      max_depth=None,
      random_state=0,
      n_jobs=-1
    )
    model.fit(Xs, y)
    return model, scaler

def evaluate(model, scaler, X, y):
    Xs = scaler.transform(featurize(X))
    preds = model.predict(Xs)
    return r2_score(y, preds), math.sqrt(mean_squared_error(y, preds))

# === Adaptive Sampling ===
def adaptive_sampling(func, bounds, X_dim, init_N, iter_N, max_samples,
                      mix_ratio, n_iter, val_N, params):
    lo, hi = bounds
    x_min = np.full(X_dim, lo)
    x_max = np.full(X_dim, hi)
    # Validation set
    U_val = x_min + np.random.rand(val_N, X_dim)*(x_max - x_min)
    X_val, y_val = simulate_samples(U_val, func)
    # Initial samples
    U0 = x_min + np.random.rand(init_N, X_dim)*(x_max - x_min)
    X_full, y_full = simulate_samples(U0, func)

    sample_counts = [len(X_full)]
    r2_tr, r2_vl, rm_vl = [], [], []

    # Iteration 0
    model, scaler = train_model(X_full, y_full, params)
    rt, _   = evaluate(model, scaler, X_full, y_full)
    rv, rmv = evaluate(model, scaler, X_val, y_val)
    r2_tr.append(rt); r2_vl.append(rv); rm_vl.append(rmv)
    print(f"Iter  0: Samples={len(X_full)}, Train R²={rt:.4f}, Val R²={rv:.4f}, Val RMSE={rmv:.4f}")

    X_err, y_err = X_full, y_full
    errors = None
    sigma = (x_max - x_min) / 10.0

    for i in range(1, n_iter+1):
        Nw = len(X_err)
        weights = np.ones(Nw)/Nw if errors is None else errors
        weights = weights.flatten()
        mix_n = int(iter_N * mix_ratio)

        existing = set(map(tuple, X_full))
        mix_samps = []
        attempts = 0
        cum_w = np.cumsum(weights)
        while len(mix_samps) < mix_n and attempts < mix_n*10:
            attempts += 1
            u = np.random.rand()
            idx = np.searchsorted(cum_w, u)
            if idx >= Nw: idx = Nw - 1
            s = np.random.multivariate_normal(X_err[idx], np.eye(X_dim)*(sigma**2))
            tup = tuple(s)
            if tup not in existing:
                mix_samps.append(s)
                existing.add(tup)
        mix_samps = np.array(mix_samps)

        # exploration sampling
        uni_n = iter_N - mix_n
        uni_samps = x_min + np.random.rand(uni_n, X_dim)*(x_max - x_min)

        # combine new samples
        new_samps = np.vstack([mix_samps, uni_samps])
        X_err, y_err = simulate_samples(new_samps, func)

        # update full dataset with cap
        X_full = np.vstack([X_full, X_err])
        y_full = np.concatenate([y_full, y_err])
        if len(X_full) > max_samples:
            X_full = X_full[-max_samples:]
            y_full = y_full[-max_samples:]
        sample_counts.append(len(X_full))

        # retrain and evaluate
        model, scaler = train_model(X_full, y_full, params)
        rt, _   = evaluate(model, scaler, X_full, y_full)
        rv, rmv = evaluate(model, scaler, X_val, y_val)
        r2_tr.append(rt); r2_vl.append(rv); rm_vl.append(rmv)

        # update errors
        preds_err = model.predict(scaler.transform(featurize(X_err)))
        pe = np.abs(preds_err - y_err)
        errors = (pe / pe.sum()).flatten()

        if i % 5 == 0:
          print(f"Iter {i:>2}: Samples={len(X_full)}, Train R²={rt:.4f}, Val R²={rv:.4f}, Val RMSE={rmv:.4f}")

    return np.array(sample_counts), np.array(r2_tr), np.array(r2_vl), np.array(rm_vl)

# === Random Sampling ===
def random_sampling(func, bounds, X_dim, sample_counts, val_N, params):
    lo, hi = bounds
    x_min = np.full(X_dim, lo)
    x_max = np.full(X_dim, hi)
    # fixed validation set
    U_val = x_min + np.random.rand(val_N, X_dim)*(x_max - x_min)
    X_val, y_val = simulate_samples(U_val, func)

    r2_tr, r2_vl, rm_vl = [], [], []
    for i, n in enumerate(sample_counts):
        U = x_min + np.random.rand(int(n), X_dim)*(x_max - x_min)
        Xs, ys = simulate_samples(U, func)
        model, scaler = train_model(Xs, ys, params)
        rt, _   = evaluate(model, scaler, Xs, ys)
        rv, rmv = evaluate(model, scaler, X_val, y_val)
        r2_tr.append(rt); r2_vl.append(rv); rm_vl.append(rmv)

        if i % 5 == 0:
          print(f"Iter {i:>2}: Samples={int(n)}, Train R²={rt:.4f}, Val R²={rv:.4f}, Val RMSE={rmv:.4f}")
    return np.array(r2_tr), np.array(r2_vl), np.array(rm_vl)

# === Main Execution ===
experiments = [
    ("Forrester (1D)",    forrester,    (0.0, 1.0),   1,  100,  30, 3000, 0.5, 100,  100),
    ("Rosenbrock (5D)",   rosenbrock5,  (-2.0, 2.0),  5,  100,  30, 3000, 0.5, 100,  250),
    ("Sobol g (10D)",     sobol_g,      (0.0, 1.0),  10,  100,  30, 3000, 0.5, 100,  500),
]
xgb_params = {
    "n_estimators": 50,
    "max_depth": 4,
    "learning_rate": 0.1,
    "objective": "reg:squarederror",
    "tree_method": "hist",
    "random_state": 0,
    "n_jobs": -1
}

for title, func, bounds, X_dim, init_N, iter_N, max_s, mix_r, n_it, val_N in experiments:
    print(f"\n=== {title} ===")
    sc, r2tr_ad, r2vl_ad, rm_ad = adaptive_sampling(
        func, bounds, X_dim, init_N, iter_N, max_s,
        mix_r, n_it, val_N, xgb_params
    )

    r2tr_rs, r2vl_rs, rm_rs = random_sampling(
        func, bounds, X_dim, sc, val_N, xgb_params
    )

    # Plot Adaptive Sampling
    plt.figure(figsize=(6,3))
    ax1 = plt.gca()
    ax1.plot(sc, r2tr_ad, 'o-', label='Train R²')
    ax1.plot(sc, r2vl_ad, 's-', label='Val R²')
    ax1.set_xlabel('Samples'); ax1.set_ylabel('R²'); ax1.set_ylim(-0.1,1.05)
    ax1.legend(loc='upper left')
    ax2 = ax1.twinx()
    ax2.plot(sc, rm_ad, '^-', color='C2', label='Val RMSE')
    ax2.set_ylabel('Val RMSE'); ax2.legend(loc='upper right')
    plt.title(f'{title} - Adaptive Sampling')
    plt.tight_layout(); plt.show()

    # Plot Random Sampling
    plt.figure(figsize=(6,3))
    ax1 = plt.gca()
    ax1.plot(sc, r2tr_rs, 'o-', label='Train R²')
    ax1.plot(sc, r2vl_rs, 's-', label='Val R²')
    ax1.set_xlabel('Samples'); ax1.set_ylabel('R²'); ax1.set_ylim(-0.1,1.05)
    ax1.legend(loc='upper left')
    ax2 = ax1.twinx()
    ax2.plot(sc, rm_rs, '^-', color='C2', label='Val RMSE')
    ax2.set_ylabel('Val RMSE'); ax2.legend(loc='upper right')
    plt.title(f'{title} - Random Sampling')
    plt.tight_layout(); plt.show()
```

```python
import numpy as np
import math
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.ensemble import RandomForestRegressor

# === Test functions ===
def forrester(x):
    x = np.array(x)
    return (6*x - 2)**2 * np.sin(12*x - 4)

def rosenbrock5(x):
    x = np.array(x)
    return np.sum(100*(x[1:] - x[:-1]**2)**2 + (x[:-1]-1)**2)

def sobol_g(x, a=None):
    x = np.array(x)
    if a is None:
        a = np.ones(x.size)
    return np.prod((np.abs(4*x-2)+a)/(1+a))

# === Utilities ===
def featurize(X):
    return np.hstack([X, np.sin(2*np.pi*X), np.cos(2*np.pi*X)])

def simulate_samples(samples, func):
    X = np.array(samples)
    y = np.array([func(x) for x in X]).reshape(-1,1)
    return X, y

def evaluate_model(model, scaler, X, y):
    Xf = featurize(X)
    Xs = scaler.transform(Xf)
    preds = model.predict(Xs)
    return r2_score(y, preds), math.sqrt(mean_squared_error(y, preds))

# === Adaptive Sampling ===
def adaptive_sampling(func, bounds, X_dim, init_N, iter_N, max_samples,
                      n_iter, val_N, rf_params, err_frac=0.5, pool_factor=20):
    lo, hi = bounds
    x_min, x_max = np.full(X_dim, lo), np.full(X_dim, hi)
    # validation set
    U_val = x_min + np.random.rand(val_N, X_dim)*(x_max - x_min)
    X_val, y_val = simulate_samples(U_val, func)
    # initial design
    U0 = x_min + np.random.rand(init_N, X_dim)*(x_max - x_min)
    X_full, y_full = simulate_samples(U0, func)
    sample_counts = [len(X_full)]
    r2_train, r2_val, rm_val = [], [], []
    for i in range(n_iter + 1):
        # train main RF
        Xf = featurize(X_full)
        scaler = StandardScaler().fit(Xf)
        Xs = scaler.transform(Xf)
        rf_main = RandomForestRegressor(oob_score=True, **rf_params)
        rf_main.fit(Xs, y_full.ravel())
        rt, _ = evaluate_model(rf_main, scaler, X_full, y_full)
        rv, rm = evaluate_model(rf_main, scaler, X_val, y_val)
        r2_train.append(rt); r2_val.append(rv); rm_val.append(rm)
        if i % 5 == 0:
          print(f"Iter {i:>2}: Samples={len(X_full)}, Train R²={rt:.4f}, Val R²={rv:.4f}, Val RMSE={rm:.4f}")
        if i == n_iter:
            break
        # error-driven residual model
        oob_pred = rf_main.oob_prediction_
        residuals = y_full.ravel() - oob_pred
        rf_resid = RandomForestRegressor(**rf_params)
        rf_resid.fit(Xs, residuals)
        # candidate pool
        pool_size = iter_N * pool_factor
        X_pool = x_min + np.random.rand(pool_size, X_dim)*(x_max - x_min)
        Xs_pool = scaler.transform(featurize(X_pool))
        # scores
        err_scores = np.abs(rf_resid.predict(Xs_pool))
        all_preds = np.stack([t.predict(Xs_pool) for t in rf_main.estimators_], axis=0)
        unc_scores = all_preds.std(axis=0)
        # select top by error and uncertainty
        n_err = int(iter_N * err_frac)
        n_unc = iter_N - n_err
        idx_err = np.argsort(-err_scores)[:n_err]
        idx_unc = np.argsort(-unc_scores)[:n_unc]
        idx_sel = np.unique(np.concatenate([idx_err, idx_unc]))
        X_new = X_pool[idx_sel]
        # evaluate and append
        Xn, yn = simulate_samples(X_new, func)
        X_full = np.vstack([X_full, Xn])
        y_full = np.vstack([y_full, yn])
        if len(X_full) > max_samples:
            X_full = X_full[-max_samples:]; y_full = y_full[-max_samples:]
        sample_counts.append(len(X_full))
    return np.array(sample_counts), np.array(r2_train), np.array(r2_val), np.array(rm_val)

# === Random Sampling Baseline ===
def random_sampling(func, bounds, X_dim, sample_counts, val_N, rf_params):
    lo, hi = bounds
    x_min, x_max = np.full(X_dim, lo), np.full(X_dim, hi)
    U_val = x_min + np.random.rand(val_N, X_dim)*(x_max - x_min)
    X_val, y_val = simulate_samples(U_val, func)
    r2_train, r2_val, rm_val = [], [], []
    for i, count in enumerate(sample_counts):
        U = x_min + np.random.rand(int(count), X_dim)*(x_max - x_min)
        X_tr, y_tr = simulate_samples(U, func)
        Xs_tr = StandardScaler().fit_transform(featurize(X_tr))
        rf = RandomForestRegressor(**rf_params)
        rf.fit(Xs_tr, y_tr.ravel())
        rt, _ = evaluate_model(rf, StandardScaler().fit(featurize(X_tr)), X_tr, y_tr)
        rv, rm = evaluate_model(rf, StandardScaler().fit(featurize(X_tr)), X_val, y_val)
        r2_train.append(rt); r2_val.append(rv); rm_val.append(rm)
        if i % 5 == 0:
          print(f"Iter {i:>2}: Samples={int(count)}, Train R²={rt:.4f}, Val R²={rv:.4f}, Val RMSE={rm:.4f}")
    return np.array(r2_train), np.array(r2_val), np.array(rm_val)

# === Updated Experiment Setup ===
if __name__ == "__main__":
    experiments = [
        ("Forrester (1D)",    forrester,    (0.0, 1.0),   1,  100,  30, 3000, 0.5, 100,  100),
        ("Rosenbrock (5D)",   rosenbrock5,  (-2.0, 2.0),  5,  100,  30, 3000, 0.5, 100,  250),
        ("Sobol g (10D)",     sobol_g,      (0.0, 1.0),  10,  100,  30, 3000, 0.5, 100,  500),
    ]
    rf_params = {"n_estimators":100, "max_depth":None, "random_state":0, "n_jobs":1}

    for title, func, bounds, X_dim, init_N, iter_N, max_s, err_frac, n_iter, val_N in experiments:
        print(f"\n=== {title} ===")
        sc, r2_tr_ad, r2_val_ad, rm_ad = adaptive_sampling(
            func, bounds, X_dim, init_N, iter_N, max_s,
            n_iter, val_N, rf_params, err_frac=err_frac, pool_factor=20
        )
        r2_tr_rs, r2_val_rs, rm_rs = random_sampling(
            func, bounds, X_dim, sc, val_N, rf_params
        )
        # Plot Adaptive
        plt.figure(figsize=(6,3))
        ax = plt.gca()
        ax.plot(sc, r2_tr_ad, 'o-', label='Adaptive Train R²')
        ax.plot(sc, r2_val_ad, 's-', label='Adaptive Val R²')
        ax.set_xlabel('Samples'); ax.set_ylabel('R²'); ax.set_ylim(-0.1,1.05)
        ax.legend(loc='upper left')
        ax2 = ax.twinx()
        ax2.plot(sc, rm_ad, '^-', label='Adaptive Val RMSE')
        ax2.set_ylabel('Val RMSE'); ax2.legend(loc='upper right')
        plt.title(f'{title} - Adaptive Sampling')
        plt.tight_layout(); plt.show()

        # Plot Random
        plt.figure(figsize=(6,3))
        ax = plt.gca()
        ax.plot(sc, r2_tr_rs, 'x--', label='Random Train R²')
        ax.plot(sc, r2_val_rs, 'd--', label='Random Val R²')
        ax.set_xlabel('Samples'); ax.set_ylabel('R²'); ax.set_ylim(-0.1,1.05)
        ax.legend(loc='upper left')
        ax2 = ax.twinx()
        ax2.plot(sc, rm_rs, 'v--', label='Random Val RMSE')
        ax2.set_ylabel('Val RMSE'); ax2.legend(loc='upper right')
        plt.title(f'{title} - Random Sampling')
        plt.tight_layout(); plt.show()
```

```python
import numpy as np
import math
import random
import pandas as pd
import matplotlib.pyplot as plt
from io import StringIO
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.ensemble import RandomForestRegressor

# === Data Parser (for CSV, included for completeness) ===
class DataParser:
    def __init__(self, data_str):
        self.data_str = data_str

    def parse(self):
        buf = StringIO(self.data_str)
        cols = buf.readline().strip().split(',')
        dt = np.genfromtxt(buf, delimiter=',')
        y = dt[:,0]
        X = dt[:,1:]
        return X, y, cols[1:]

# === Test Functions ===
def forrester(x):
    x = np.array(x)
    return (6*x - 2)**2 * np.sin(12*x - 4)

def rosenbrock5(x):
    x = np.array(x)
    return np.sum(100*(x[1:]-x[:-1]**2)**2 + (x[:-1]-1)**2)

def sobol_g(x, a=None):
    x = np.array(x)
    if a is None:
        a = np.ones(x.size)
    return np.prod((np.abs(4*x-2)+a)/(1+a))

# === Utilities ===
def featurize(X):
    return np.hstack([X, np.sin(2*np.pi*X), np.cos(2*np.pi*X)])

def simulate_samples(X, func):
    y = np.array([func(x) for x in X]).reshape(-1,1)
    return X, y

def surrogate_model(X, y, rf_params):
    Xf = featurize(X)
    scaler = StandardScaler().fit(Xf)
    Xs = scaler.transform(Xf)
    model = RandomForestRegressor(oob_score=True, **rf_params)
    model.fit(Xs, y.ravel())
    return model, scaler

def evaluate_model(model, scaler, X, y):
    Xs = scaler.transform(featurize(X))
    preds = model.predict(Xs)
    return r2_score(y, preds), math.sqrt(mean_squared_error(y, preds))

def samples_from_mixture(X_existing, weights, means, sigma, N):
    samples = []
    existing = set(map(tuple, X_existing))
    dim = means.shape[1]
    while len(samples) < N:
        idx = np.random.choice(len(weights), p=weights)
        s = np.random.multivariate_normal(means[idx], np.eye(dim)*(sigma[idx]**2))
        if tuple(s) not in existing:
            samples.append(s)
            existing.add(tuple(s))
    return np.array(samples)

# === Adaptive Sampler ===
def adaptive_sampler(func, bounds, X_dim, init_N, iter_N, max_samples,
                     n_iter, val_N, rf_params,
                     err_frac_max=0.5, pool_factor=20, exploration_iters=3):
    lo, hi = bounds
    x_min, x_max = np.full(X_dim, lo), np.full(X_dim, hi)
    # validation set
    X_val = x_min + np.random.rand(val_N, X_dim)*(x_max - x_min)
    X_val, y_val = simulate_samples(X_val, func)
    # initial design
    X_full = x_min + np.random.rand(init_N, X_dim)*(x_max - x_min)
    X_full = np.unique(X_full, axis=0)
    X_full, y_full = simulate_samples(X_full, func)
    # records
    sample_counts, r2_tr, r2_vl, rm_vl = [], [], [], []

    for i in range(n_iter+1):
        model, scaler = surrogate_model(X_full, y_full, rf_params)
        rt, _ = evaluate_model(model, scaler, X_full, y_full)
        rv, rm = evaluate_model(model, scaler, X_val, y_val)
        sample_counts.append(len(X_full)); r2_tr.append(rt); r2_vl.append(rv); rm_vl.append(rm)
        if i % 5 == 0:
          print(f"Iter {i:>2}: Samples={len(X_full)}, Train R²={rt:.4f}, Val R²={rv:.4f}, Val RMSE={rm:.4f}")
        if i == n_iter:
            break

        # dynamic err_frac
        if i < exploration_iters:
            err_frac = 0.0
        else:
            err_frac = err_frac_max * ((i-exploration_iters)/(n_iter-exploration_iters))

        # residual model
        residuals = y_full.ravel() - model.oob_prediction_
        rf_resid = RandomForestRegressor(**rf_params)
        rf_resid.fit(scaler.transform(featurize(X_full)), residuals)

        # candidate pool
        pool_size = iter_N * pool_factor
        X_pool = x_min + np.random.rand(pool_size, X_dim)*(x_max - x_min)
        Xs_pool = scaler.transform(featurize(X_pool))

        # scores
        err_scores = np.abs(rf_resid.predict(Xs_pool))
        all_preds = np.stack([t.predict(Xs_pool) for t in model.estimators_], axis=0)
        unc_scores = all_preds.std(axis=0)

        # select
        n_err = int(iter_N * err_frac); n_unc = iter_N - n_err
        idx_err = np.argsort(-err_scores)[:n_err]
        idx_unc = []
        for idx in np.argsort(-unc_scores):
            if idx not in idx_err and len(idx_unc) < n_unc:
                idx_unc.append(idx)
            if len(idx_unc) >= n_unc:
                break
        idx_sel = np.concatenate([idx_err, idx_unc])
        X_new = X_pool[idx_sel]

        # append new
        Xn, yn = simulate_samples(X_new, func)
        X_full = np.vstack([X_full, Xn]); y_full = np.vstack([y_full, yn])
        # cap
        if len(X_full) > max_samples:
            X_full = X_full[-max_samples:]; y_full = y_full[-max_samples:]

    return np.array(sample_counts), np.array(r2_tr), np.array(r2_vl), np.array(rm_vl)

# === Random Sampler ===
def random_sampler(func, bounds, X_dim, sample_counts, val_N, rf_params):
    lo, hi = bounds
    x_min, x_max = np.full(X_dim, lo), np.full(X_dim, hi)
    X_val = x_min + np.random.rand(val_N, X_dim)*(x_max - x_min)
    X_val, y_val = simulate_samples(X_val, func)
    r2_tr, r2_vl, rm_vl = [], [], []
    for i, n in enumerate(sample_counts):
        X_tr = x_min + np.random.rand(int(n), X_dim)*(x_max - x_min)
        _, y_tr = simulate_samples(X_tr, func)
        model, scaler = surrogate_model(X_tr, y_tr, rf_params)
        rt, _ = evaluate_model(model, scaler, X_tr, y_tr)
        rv, rm = evaluate_model(model, scaler, X_val, y_val)
        r2_tr.append(rt); r2_vl.append(rv); rm_vl.append(rm)
        if i % 5 == 0:
          print(f"Rand Iter {i:>2}: Samples={n}, Val R²={rv:.4f}, RMSE={rm:.4f}")
    return np.array(r2_tr), np.array(r2_vl), np.array(rm_vl)

# === Main Execution & Plot ===
if __name__ == "__main__":
    experiments = [
        ("Forrester (1D)", forrester, (0.0, 1.0), 1, 100, 30, 3000, 0.5, 100, 100),
        ("Rosenbrock (5D)", rosenbrock5, (-2.0, 2.0), 5, 100, 30, 3000, 0.5, 100, 250),
        ("Sobol g (10D)", sobol_g, (0.0, 1.0), 10, 100, 30, 3000, 0.5, 100, 500),
    ]
    rf_params = {"n_estimators":100, "max_depth":None, "random_state":0, "n_jobs":1}

    for title, func, bounds, X_dim, init_N, iter_N, max_s, err_frac, n_iter, val_N in experiments:
        print(f"\n=== {title} ===")
        sc, r2_tr_ad, r2_vl_ad, rm_ad = adaptive_sampler(
            func, bounds, X_dim, init_N, iter_N, max_s,
            n_iter, val_N, rf_params, err_frac_max=err_frac, pool_factor=20, exploration_iters=3
        )
        r2_tr_rs, r2_vl_rs, rm_rs = random_sampler(
            func, bounds, X_dim, sc, val_N, rf_params
        )

        # Adaptive plot
        fig, ax1 = plt.subplots(figsize=(6,3))
        ax1.plot(sc, r2_tr_ad, 'o-', label='Adaptive Train R²')
        ax1.plot(sc, r2_vl_ad, 's-', label='Adaptive Val R²')
        ax1.set_xlabel('Samples'); ax1.set_ylabel('R²'); ax1.set_ylim(-0.1,1.05)
        ax1.legend(loc='upper left')
        ax2 = ax1.twinx()
        ax2.plot(sc, rm_ad, '^-', label='Adaptive Val RMSE'); ax2.set_ylabel('RMSE')
        ax2.legend(loc='upper right')
        plt.title(f'{title} - Adaptive'); plt.tight_layout(); plt.show()

        # Random plot
        fig, ax1 = plt.subplots(figsize=(6,3))
        ax1.plot(sc, r2_tr_rs, 'x--', label='Random Train R²')
        ax1.plot(sc, r2_vl_rs, 'd--', label='Random Val R²')
        ax1.set_xlabel('Samples'); ax1.set_ylabel('R²'); ax1.set_ylim(-0.1,1.05)
        ax1.legend(loc='upper left')
        ax2 = ax1.twinx()
        ax2.plot(sc, rm_rs, 'v--', label='Random Val RMSE'); ax2.set_ylabel('RMSE')
        ax2.legend(loc='upper right')
        plt.title(f'{title} - Random'); plt.tight_layout(); plt.show()
```
